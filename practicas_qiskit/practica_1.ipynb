### Introducción a Qiskit Práctica I

Para crear un circuito debemos importar el modulo respectivo con la clase QuantumCircuit:

```python
from qiskit import QuantumCircuit
```

```python
#Creamos una instancia de la clase QuantumCircuit
qc=QuantumCircuit(3) #Este circuito tiene 3 q-bits
```

```python
#Reseteamos cada q-bit
qc.reset(0)
qc.reset(1)
qc.reset(2)
```

```python
#Establecemos una puerta Hadamard en el primer q-bit
qc.h(0)
```

```python
#Podemos agregar puertas CNOT a los q-bits de nuestro circuito
qc.cx(0,1)
qc.cx(0,2)
#Medimos los q-bits
qc.measure_all()
```

```python
#Mostramos la representación gráfca del circuito:
qc.draw(output="mpl",style="iqp")
```
